```python
import requests
from datetime import datetime, timedelta

ORG = "openclaw"
# Define la fecha límite (1 mes atrás desde hoy 15 de abril de 2026)
since_date = (datetime.now() - timedelta(days=30)).isoformat()

def get_active_repos(org):
    url = f"https://api.github.com/orgs/{org}/repos"
    # Nota: Asegúrate de usar un Token de GitHub si tienes problemas de rate limit
    response = requests.get(url)
    repos = response.json()
    
    active_repos = [
        r['clone_url'] for r in repos 
        if r['pushed_at'] > since_date and not r['fork']
    ]
    return active_repos

repo_list = get_active_repos(ORG)
print(f"Repositorios encontrados para analizar: {len(repo_list)}")
```
